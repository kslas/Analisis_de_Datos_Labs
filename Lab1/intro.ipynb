```python
#importacion librerias 
import pandas as pd
import matplotlib.pyplot as plt


plt.style.use('ggplot')

#url dataset
url = 'https://raw.githubusercontent.com/cs109/2014_data/master/diamonds.csv'

diamonds = pd.read_csv(url, sep = ',', index_col=0)
diamonds.head()

```

```python
diamonds.describe()
```

```python
diamonds['price'].hist(bins=50, color = 'green')
plt.title('Distribution of Price')
plt.xlabel('Price')
```

```python
# Try changing bin size from 20 to 500. What do you notice? 
diamonds['carat'].hist(bins=20, color = 'green', figsize=(6, 4))
plt.title('Distribution of weights in carats')
plt.xlim(0, 3)
plt.xlabel('Weight in Carats')
```

```python
diamonds['price'].plot(kind='kde', color = 'green')
plt.title('Distribution of Price')
```

```python
diamonds.plot(x='carat', y='price', kind = 'scatter', color = 'green', alpha = 1)
```

```python
plt.scatter(diamonds['carat'], diamonds['price'], color = 'green', alpha = 0.05)
plt.xlabel('Carat')
plt.ylabel('Price')
```

```python
diamonds.groupby('color').plot(x='carat', y='price', kind = 'scatter', color = 'green', alpha = 1)
```

```python
diamonds.groupby('clarity').plot(x='carat', y='price', kind = 'scatter', color = 'green', alpha = 1)
```
